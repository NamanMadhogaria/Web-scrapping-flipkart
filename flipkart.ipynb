```python
import requests
from bs4 import BeautifulSoup
product=[]
import pandas as pd
price=[]
desc=[]
for i in range (2,5):
    url="https://www.flipkart.com/search?q=sports+shoes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
    box=soup.find("div",class_="DOjaWF gdgoEp")
    names=box.find_all("div",class_="syl9yP")
    for i in names:
        name=i.text
        product.append(name)
    #print(len(product))
    names=box.find_all("div",class_="Nx9bqj")
    for i in names:
        prices=i.text
        price.append(prices)
    #print(len(price))
    df=pd.DataFrame({"Company":product,"prices":price})
    print(df)
    df.to_csv("Downloads/sports.csv")
```
